# Exploratory data analysis in Python

## 1. Importing the libraries 

```python
import pandas as pd 
import numpy as np 

import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline 
sns.set(color_codes=True)
```

## 2. Loading dara

## 4. Dropping columns for instance 

```python
df.columns
```

```python
df =df.drop(['Engine Fuel Type', 'Market Category','Vehicle Size', 'Vehicle Style','Number of Doors', 'Popularity'], axis=1)

df
```

## 5. Renaming the columns 

```python
df.columns
```

```python
rename_dic={'Engine HP':'HP','Engine Cylinders':'Cylinders', 'Transmission Type':'Transmission', 'highway MPG':'MPG-H', 'city mpg':'MPG-C' }

df = df.rename(columns= rename_dic)

df
```

## 6. Dropping the duplicate rows 

```python
df.shape
```

```python
duplicate_df = df[df.duplicated()]

duplicate_df
```

```python
df = df.drop_duplicates()

df
```

## 7. Dropping the missing or null values

```python
df.isnull().sum()
```

```python
df = df.dropna()

df.isnull().sum()
```

## 8. Detecting Outliers

```python
df
```

```python
df.columns

df.info()
```

```python
sns.boxplot(x= df['HP'])
```

```python
sns.boxplot(x= df['Cylinders'])
```

```python
sns.boxplot(x= df['MPG-H'])
```

```python
sns.boxplot(x= df['MPG-C'])
```

```python
# MSRP

sns.boxplot(x= df['MSRP'])
```

```python
q1=df.quantile(0.25)
q3=df.quantile(0.75)

IQR = q3-q1

IQR
```

```python
df = df[~((df < (q1-1.5*IQR)) | ((df>(q3+1.5*IQR)))).any(axis=1)]

df
```

## 9. Plot different features against one another

### Histogram

```python
df.columns

df.info()
```

```python
df['Make'].value_counts()
```

```python
df['Make'].value_counts().nlargest(40)
```

```python
df['Make'].value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))

plt.title('Number of cars by make of TOP 40')
plt.ylabel('Number of Cars')
plt.xlabel('Make')
```

### Heat MAPS 

```python
# Finding the relations between the variables 

plt.figure(figsize=(20,10))

c= df.corr()
sns.heatmap(c, cmap='BrBG', annot=True)
c
```
